```python
## plotting using matplotlib

## plt scatter
## c is used to specify color here g is green 
## if we dont write this  (%matplotlib inline) above then we need to write plt.show() 

plt.scatter(x,y,c='g')
plt.xlabel('X axis')         ## xlabel, ylabel are inbuilt funct
plt.ylabel('Y axis')
plt.title("Graph in 2D")     ## giving description to the graph (title)
plt.savefig("Graph_1.png")   ## saving the graph

```

```python
##creating Subplots 
## format is:
## subplot(no.of rows, no.of cols, index, **kwargs)
   ## subplot(pos, **kwargs)


plt.subplot(2,2,1)
plt.plot(x,y,'g*--')
plt.subplot(2,2,2)
plt.plot(x,y,'co')
plt.subplot(2,2,3)
plt.plot(x,y,'y*')
plt.subplot(2,2,4)
plt.plot(x,y,'r*--')
```

```python
#Subplot
#Compute x and y coordinates for points on sine n cosine curves

x=np.arange(0,5*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)

# Set up a subplot grid that has height 2 and width 1
# And set the first such subplot as active.
plt.subplot(2,1,1)

#Make the first plot 
plt.plot(x,y_sin,'r--')
plt.title('Sine')

# Set the second subplot as active and the second plot.
plt.subplot(2,1,2)
plt.plot(x,y_cos,'y*--')
plt.title('Cosine')

#Show the fig 
plt.show()
```

## BAR PLOT

```python
## BAR PLOT 

x1=[2,8,10]
y1=[11,16,9]

x2=[3,9,11]
y2=[6,15,7]
plt.bar(x1,y1,align='center')
plt.bar(x2,y2,color='g')
plt.title("BAR GRAPH")
plt.xlabel("X-AXIS")
plt.ylabel("Y-AXIS")
```

## HISTOGRAMS

```python
## values are plotted in x axis and 
## the density is shown by the y-axis

 
a=np.array([22,87,5,56,4,73,54,11,25,7,31,26])
plt.hist(a)
plt.title("HISTOGRAM")
plt.show()
```

## BOX PLOT USING MATPLOTLIB

```python
## vert refers to vertical and the plot is vertical..if we put vert=False,then the plot will become horizontal 
## patch_artist is used for giving colors, if we set the values as False the color will be removed.


data = [np.random.normal(0 ,std ,100) for std in range (1,4)]

# rectangular box plot 
plt.boxplot(data,vert=True,patch_artist=True);
```

```python
data = [np.random.normal(0 ,std ,100) for std in range (1,4)]

# rectangular box plot 
plt.boxplot(data,vert=False,patch_artist=False);
```
