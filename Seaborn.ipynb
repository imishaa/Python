```python
import seaborn as sns
```

## Correlation with Heatmap





A correlation heatmap uses colored cells,typically in a monochromatic scale, to show a 2D correlation matrix(table) between two discete dimension or event types. It is very important in Feature Selection.

```python
## to find corr the values must be integer or float 

df.corr()
```

```python
sns.heatmap(df.corr())
```

## JoinPlot

A joinplot allows to study the relationship between 2 numeric variables. The central chart display their correlation.It is usually a scatterplot,a hexbin plot,a 2D histogram or a 2D density plot.

## Univariate Analysis

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')..........##reg is regression

##this line over graph is probability distribution graph line 
```

## Pairplot

A "Pairplot" is known as scatterplot,in which one variabe in the same data row is matched with another variable's value,like this: Pair plots are just elaborations on this,showing all variables paired with all the other variables.

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

## Distplot


Distplot helps us to check the distribution of the columns feature

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
```

```python
#count plot
sns.countplot('smoker',data=df)
```

```python
sns.boxplot(x='smoker',y='total_bill',data=df)
```

```python
sns.boxplot(x='day',y='total_bill', data=df, palette='rainbow')
```

```python
sns.boxplot(x='total_bill',y='day',hue='smoker',data=df)
```

# Violin Plot


  Violit plot helps us to see both distribution of data in terms of Kernel Density estimation n the boxplot.

```python
sns.violinplot(x='total_bill',y='day',data=df,palette='rainbow')
```
